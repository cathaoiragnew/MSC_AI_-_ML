# **Used Resources** 
**********************************

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html


https://stackoverflow.com/questions/53519823/confidence-interval-in-python-dataframe

```
#stats_score = df_all.groupby(['Controller'])['Score'].agg(['mean', 'min', 'max', 'std'])
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"


stats_score = df_all.groupby(['Controller'])['Score'].agg(['mean', 'count', 'std'])
print(stats_score)
print('-'*55)

ci95_hi = []
ci95_lo = []


for i in stats_score.index:
    m, c, s = stats_score.loc[i]
    ci95_hi.append(m + 1.96*s/math.sqrt(c))
    ci95_lo.append(m - 1.96*s/math.sqrt(c))

stats_score['ci95_lo'] = ci95_lo
stats_score['ci95_hi'] = ci95_hi


stats_score['min'] = df_all.groupby(['Controller'])['Score'].min()
stats_score['max'] = df_all.groupby(['Controller'])['Score'].max()
stats_score['median'] = df_all.groupby(['Controller'])['Score'].median()
stats_score['var'] = df_all.groupby(['Controller'])['Score'].var()

display(stats_score)
```

Median 95 % CI

```

stats_score_med = df_all.groupby(['Controller'])['Score'].agg(['median', 'count', 'std'])
print(stats_score_med)
print('-'*55)

ci95_hi_m = []
ci95_lo_m = []


for i in stats_score.index:
    m, c, s = stats_score_med.loc[i]
    ci95_hi_m.append(m + 1.96*s/math.sqrt(c))
    ci95_lo_m.append(m - 1.96*s/math.sqrt(c))

stats_score_med['ci95_lo'] = ci95_lo_m
stats_score_med['ci95_hi'] = ci95_hi_m

display(stats_score_med)
```

```
stats_time = df_all.groupby(['Controller'])['Time'].agg(['mean', 'count', 'std'])
print(stats_time)
print('-'*55)

ci95_hi_time = []
ci95_lo_time = []


for i in stats_time.index:
    m, c, s = stats_time.loc[i]
    ci95_hi_time.append(m + 1.96*s/math.sqrt(c))
    ci95_lo_time.append(m - 1.96*s/math.sqrt(c))

stats_time['ci95_lo_time'] = ci95_lo_time
stats_time['ci95_hi_time'] = ci95_hi_time


stats_time['min'] = df_all.groupby(['Controller'])['Time'].min()
stats_time['max'] = df_all.groupby(['Controller'])['Time'].max()
stats_time['median'] = df_all.groupby(['Controller'])['Time'].median()

display(stats_time)
```
