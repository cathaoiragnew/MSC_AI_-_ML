```python
class MatrixOperations:
    
    def __init__(self):
        pass
    
    # Getting size of a m,n function 
    def matrix_size(self, matrix):
        """Function that takes a matrix in, in the form of a list of lists to represent the matrix
           and returns the dimension of the matrix as a tuple. 
           Returns dimension of a mxn matrix as (m,n), where m = rows , n = cols. """
        try:
            
            rows = len(matrix)
            cols = len(matrix[0])

            return (rows,cols)
        
        except Exception as err:
                print("Exception was thrown.")
                raise


    
    # Gets the sum of 2 matrices
    def matrix_sum(self, matrix1, matrix2): 
        """Function that takes in 2 matrices, and sums them together using the proper mathematical techniques of matrix
           addition. The dimensions of the 2 matrices should be equal, as this is a requirement for matrix addition."""
    
        #first checking there dimensions are the same 
        if self.matrix_size(matrix1) == self.matrix_size(matrix2):
            
            try:
        
                # since dimensions of resulting matrix of adding two matrices will be equal to dimensions
                # of matrix1 = martix 2. New result matrix has rows & cols = matrix 1 rows and cols 
                
                rows = self.matrix_size(matrix1)[0]
                cols = self.matrix_size(matrix1)[1]

                # Below is creating an empty list called results the same dimensions of matrix 1 
                # this will hold the values for summing matrix, needs to be initialized before adding to it 
    
                  
                result = [ [ 0 for col in range(cols) ] for row in range(rows) ]
                
                # need to iterate for both rows & columns, will use i & j
                # as this is the standard maths notion for indexing matrics
                
                # this will iterate of the rows 
                for i in range(rows):
                # below iterates over columns 
                    for j in range(cols):
                        #calculating the sum of the matrices
                        result[i][j] = matrix1[i][j] + matrix2[i][j]
                        
                # printing out the results
                for r in result:
                    print(r)
            
            except Exception as err:
                print("Exception was thrown.")
                raise

        else: print("Matrix dimensions don't match.")  
                  
        
    #subtraction of 2 matrices 
    def matrix_subtract(self, matrix1, matrix2):
        """Function that takes in 2 matrices, and subtracts them using the proper mathematical techniques of matrix
           subtraction. The dimensions of the 2 matrices should be equal, as this is a requirement for matrix addition."""
    
        #first checking there dimensions are the same 
        if self.matrix_size(matrix1) == self.matrix_size(matrix2):
            
            try:
                
                # since dimensions of resulting matrix of subtracting two matrices will be equal to dimensions
                # of matrix1 = martix 2. New result matrix has rows & cols = matrix 1 rows and cols 
                
                rows = self.matrix_size(matrix1)[0]
                cols = self.matrix_size(matrix1)[1]

               # Below is creating an empty list called results the same dimensions of matrix 1 
                # this will hold the values for subtracting matrix, needs to be initialized before adding to it 

                result = [ [ 0 for col in range(cols) ] for row in range(rows) ]

                # need to iterate for both rows & columns, will use i & j 
                # as this is the standard maths notion for indexing matrics 
                
                # this will iterate of the rows 
                for i in range(rows):
                # below iterates over columns
                    for j in range(cols):
                        # subtracting the matrices
                        result[i][j] = matrix1[i][j] - matrix2[i][j]

                # printing out the results       
                for r in result:
                    print(r)
                    
            except Exception as err:
                print("Exception was thrown.")
                raise
                   
           
        else: print("Matrix dimensions don't match.")  
       

    # function that calulcates matrix multiplication of 2 matrices
    def matrix_multiple(self, matrix1, matrix2):
        """Function that takes in 2 matrices, and multiples them using the proper mathematical techniques of matrix
           multiplication. Matrix multiplaction requires the dimensions of the matrices to be multipled to be in the form:
           Matrix 1 dimension = M x N    Matrix 2 dimension = N x T , where M,N,T are some number.    
           This is a requirement for matrix multiplication."""
    
    
        #first checking that matrix1 is mxn dimensions and matrix2 is nxt , where m,n,t can be any number 
        if self.matrix_size(matrix1)[1] == self.matrix_size(matrix2)[0]:
            
            try:

                # for matrix multiplcation the result from a mxn by nxt , will result in a mxt matrix
                # in english, the length of columns of matrix 1 have to match the length rows of matrix2 , 
                # the resulting matrix will be of dimension; matrix 1 rows and matrix 2 columns

                #going to calculate rows & cols with dimensions mxt  
                rows = self.matrix_size(matrix1)[0]
                cols = self.matrix_size(matrix2)[1]


                # this is creating an empty list called results with dimensions rows/cols from above    
                # initialize result matrix of dimensions row,cols which = m x t   
                
                result = [ [ 0 for col in range(cols) ] for row in range(rows) ]

                # need to iterate for rows & columns of matrix 1 and rows & cols for matrix2
                # however since len(cols) of matrix 1, needs to be equal to len(rows) of matrix 2 , 
                # they share iterator, so we need 3 iterators

                #will use i & j & k , as iterators

                # below is indexing for rows of matrix 1
                for i in range(self.matrix_size(matrix1)[0]):
                    
                # below handles columns of matrix 2 
                    for j in range(self.matrix_size(matrix2)[1]):
                        
                        # finally rows of matrix 2 / but this could also be cols of martix 1  
                        for k in range(self.matrix_size(matrix2)[0]):
                            result[i][j] += matrix1[i][k] * matrix2[k][j]

                # just going to return it like this, as it can them be put directly into plotting function
                return result
            
            except Exception as err:
                print("Exception was thrown.")
                raise
                

        else: print("Matrix dimensions don't match.") 
            
      
    # calculate the determinent of a 2x2 matrix 
    def matrix_det(self, matrix):
        """Function that takes in a matrix, in the form of a list of lists to represent the matrix
           and calculates the determinant of the matrix ."""

        #first checking that matrix is 2x2 dimension 
        if self.matrix_size(matrix) == (2,2):
            
            try:
                
                # below is the forumals to calculate the determinant of a 2x2 matrix
                
                result = matrix[0][0] * matrix[1][1]  - (matrix[1][0] * matrix[0][1])

                return result
            
            except Exception as err:
                print("Exception was thrown.")
                raise

        else: print("Matrix dimensions don't match.")    
            


    # calculate the inverse of a 2x2 matrix 
    def matrix_inverse(self, matrix):
        """Function that takes in a matrix, in the form of a list of lists to represent the matrix
           and calculates the inverse of the matrix ."""
        
        #first checking that matrix is 2x2 dimension 
        if self.matrix_size(matrix) == (2,2):
            
            try:

                rows = self.matrix_size(matrix)[0]
                cols = self.matrix_size(matrix)[1]


                # so first intitalize a results matrix to avoid anything from being overwritten and lost
                
                # this is creating an empty list called results with dimensions rows/cols from above    
                # initialize result matrix of dimensions row,cols   
                
                result = [ [ 0 for col in range(cols) ] for row in range(rows) ]

                
                #going to calcualte the det(matrix) & then put 1/det(matrix) as this is needed to calculate inverse
                # if det(matrix) = 0 , then the matrix is not invertible 
                
                determin = self.matrix_det(matrix)
                inv_determin = 1/determin
                
                # how to inverse a 2x2 matrix, swap diagonals, negate non-diagonals & multiple matrix by 1/det(matrix)
                
                # so first if i = j , swap i & j (swapping diagonals) , then i != j , just negate it 

                # below is indexing for rows of matrix 
                for i in range(rows):
                    # below handles columns of matrix   
                    for j in range(cols):
                        if i == j:
                            result[i][j] = inv_determin * matrix[i-1][j-1]
                        if i != j:
                            result[i][j] = inv_determin * -(matrix[i][j])

                for r in result:
                    print(r)
                
                # exception to check for division by zero, ie matrix is invertible
            except ZeroDivisionError as zero_err:
                print("There a division by zero error, so matrix is invertible, as det(matrix)=0 ")
                raise
            
            except Exception as err:
                print("Exception was thrown.")
                raise
                
                
        else: print("Matrix dimensions don't match.")


    # plots a 3x3 dimension matrix
    def matrix_3d_plot_3x3(self, matrix, x_lim = (-5,5) , y_lim = (-5,5) , z_lim = (-5,5)):
        """Function that takes in a matrix, in the form of a list of lists to represent the matrix
           and plots the basis vectors. Input needs to be of dimension 3x3."""
        
        #check that matrix is of dimension 3x3
        if self.matrix_size(matrix) == (3,3):
            
            try:
 
                # below is just needed to run quiver & plot
                origin = [0,0,0]
        
                X, Y, Z = zip(origin,origin,origin)

                # we want to access columns to see how basic vectors transform
                
                U, V, W = matrix[0][0] , matrix[1][0], matrix[2][0]
        
                M, N, S =  matrix[0][1] , matrix[1][1], matrix[2][1]
        
                Q, R, T =  matrix[0][2] , matrix[1][2], matrix[2][2]

                fig = plt.figure()
                ax = fig.add_subplot(111, projection='3d')
                ax.set_xlim3d(x_lim)
                ax.set_ylim3d(y_lim)
                ax.set_zlim3d(z_lim)

                ax.set_xlabel('x', fontsize=30, rotation=150)
                ax.set_ylabel('y', fontsize=30)
                ax.set_zlabel('z', fontsize=30)

                plt.quiver(X,Y,Z,U,V,W)
                plt.quiver(X,Y,Z,M,N,S)
                plt.quiver(X,Y,Z,Q,R,T)
                plt.show()
                
            except Exception as err:
                print("Exception was thrown.")
                raise
                
        else: print("Only plots a 3x3 dimension matrix")
```

### 1. Getting the size of a function and returning the dimensions as a tuple

Using the length of the matrix, will give us the number of rows. Cheching the length of any of the lists (which are the rows) will give us the number of columns

```python
# Getting size of a m,n function 
def matrix_size(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    return (rows,cols)
```

### 2. Returning the sum of two matrices

Making use of the function from above, matrix_size the size of matrices are checked. As matrix addition is element by element addition, the below code performs this looping over the rows and columnsand adding the relevent elements. To prevent any matrix from being overwritten, I created a new matrix filled with zero's of the same dimensions of the resulting matrix of the addition. This needed to be initalized before the the addition. 

```python
def matrix_sum(matrix1, matrix2):
    #first checking there dimensions are the same 
    if matrix_size(matrix1) == matrix_size(matrix2):
        
        # since dimensions of matrix1 = martix 2 , going to calculate rows & cols from them 
        rows = matrix_size(matrix1)[0]
        cols = matrix_size(matrix1)[1]
        
        # this is creating an empty list called results the same dimensions of matrix 1 
        # this will hold the values sum/subtract, needs to be initialized before adding to it 
        # result = [ [None] * cols ] * (rows) 
        
          
        result = [ [ 0 for col in range(cols) ] for row in range(rows) ]
       
        # need to iterate for both rows & columns, will use i & j , as this is the standard maths notion for indexing matrics 
        for i in range(len(matrix1)):
        # below handles the columns 
            for j in range(len(matrix1[0])):
                result[i][j] = matrix1[i][j] + matrix2[i][j]
    
        for r in result:
            print(r)
          
    else: print("Matrix dimensions don't match.")    
```

### 1. Function to calculate deteminant of a matrix
Making use of the function from above, matrix_size, the size of matrices are checked. For matrix determinats of a 2x2 matrix, it is calculated as followed: matrix[0][0] * matrix[1][1]  - (matrix[1][0] * matrix[0][1]), as this is just the rules of maths.

```python
# calculate the deteremint of a 2x2 matrix 
def matrix_det(matrix):
    
    #first checking that matrix is 2x2 dimension 
    if matrix_size(matrix) == (2,2):

        # just calculating determinant for a 2x2 is the below formula
        result = matrix[0][0] * matrix[1][1]  - (matrix[1][0] * matrix[0][1])
        
        return result
          
    else: print("Matrix dimensions don't match.")         
```

### 2. Function to calculate inverse of a matrix
The inverse of a 2x2 matrix is simply calculated as: 1/det(matrix) * swap the diagonals & negate the off diagonals. Making use of the function from above, matrix_size, the size of matrices are checked, it only allows matrices of dimension 2x2.The resulting matrix will be of the same dimesion of the original matrix. An empty results matrix of 2x2 dimensions is created to avoid overwriting the original matrix. With a little thought it was clear how to swap the diagonals: when i = j = 0, 
we need to swap this with i = j = 1, ie last entry of the last list, so we can just use i-1, as when i = 0  [-1] , which accesses the last index. When i = 1 , i-1 = 0 , which is the first entry of the first list. Also multiplied 1/det(metrix) when swapping & negating off diagonals to kill 2 birds with 1 stone.

```python
# calculate the inverse of a 2x2 matrix 
def matrix_inverse(matrix):
    #first checking that matrix is 2x2 dimension 
    if matrix_size(matrix) == (2,2):
    
        rows = matrix_size(matrix)[0]
        cols = matrix_size(matrix)[1]
        
            
        # so first intitalize a results matrix to avoid anything from being overwritten and lost 
        # this is creating an empty list called results with dimensions rows/cols from above    
        # initialize result matrix of dimensions row,cols    
        result = [ [ 0 for col in range(cols) ] for row in range(rows) ]
        
        #going to calcualte the det(matrix) 
        determin = matrix_det(matrix)
        
        inv_determin = 1/determin
    
        # so first if i = j , swap i & j (swapping diagonals) , then i != j , just negate it 
        
        # below is indexing for rows of matrix 
        for i in range(rows):
            # below handles columns of matrix   
            for j in range(cols):
                if i == j:
                    result[i][j] = inv_determin * matrix[i-1][j-1]
                if i != j:
                    result[i][j] = inv_determin * -(matrix[i][j])
                             
        for r in result:
            print(r)
          
    else: print("Matrix dimensions don't match.")   
```

### Plotting 3D, 3x3 matrices / Vectors 
Imported hints packages. Function only plots 3,3 matricies. So need to perform 3 different plots, X,Y,Z will always remain at the origin. Howevever to correctly plot the lines I broke the matrix up into the 3 column vectors in essence. First column is i vector, second column is j, third column is k. Added the x,y,z limits as a parameter in the function, so if need, we could see increase the axis scalings and see the arrows.

```python
#import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

def matrix_3d_plot_3x3(matrix, x_lim = (-5,5) , y_lim = (-5,5) , z_lim = (-5,5)):
    
    if matrix_size(matrix) == (3,3):

        origin = [0,0,0]
        X, Y, Z = zip(origin,origin,origin)

        U, V, W = matrix[0][0] , matrix[1][0], matrix[2][0]
        
        M, N, S =  matrix[0][1] , matrix[1][1], matrix[2][1]
        
        Q, R, T =  matrix[0][2] , matrix[1][2], matrix[2][2]

        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.set_xlim3d(x_lim)
        ax.set_ylim3d(y_lim)
        ax.set_zlim3d(z_lim)

        ax.set_xlabel('x', fontsize=30, rotation=150)
        ax.set_ylabel('y', fontsize=30)
        ax.set_zlabel('z', fontsize=30, rotation=0)

        plt.quiver(X,Y,Z,U,V,W)
        plt.quiver(X,Y,Z,M,N,S)
        plt.quiver(X,Y,Z,Q,R,T)
        plt.show()
        
    else: print("Only plots a 3x3 dimension matrix")
```

```python
ident3 =  [ [1,0,0] ,
            [0,1,0] , 
            [0,0,1]  ]


matrix_3d_plot_3x3(ident3)
```
