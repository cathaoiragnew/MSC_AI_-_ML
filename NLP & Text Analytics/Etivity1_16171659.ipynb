```
import re
import csv 

!wget https://gist.githubusercontent.com/ajoorabchi/eac194a79dd26de8864f9206b7842ff1/raw/8ea1d8d5f74b5b2724e378b43d4df6094990c7db/Eircode%2520Routing%2520Key%2520Boundaries.csv

filePath = "/content/Eircode Routing Key Boundaries.csv"

with open(filePath, 'r') as f:
  reader = csv.reader(f)
  KeyRouteList = list(map(tuple, reader))

print(KeyRouteList)

def eircodeValidator(eircode):
  # only allow A-Z & 0-9 characters
  reg = re.compile(r'[A-Z0-9]')
  eircode_res = reg.match(eircode)
    
  if bool(eircode_res) == 1:
    # printing the eircode if it was of valid characters
    print("Eircode:" , eircode )

  # going to loop through and see if first 3 characters of Eircode match
  # with any keys in the list of tuples
    for i in range(len(KeyRouteList)):
      #getting all the keys 
      key = KeyRouteList[i]
      #checking if any of the keys match with first 3 characters of Eircode
      # if theres a match, will print out the district, ie Limerick , Ennis
      if eircode[0:3] == key[0]:
        print("Valid Eircode, Routing key:" , eircode[0:3], "," , "Unique Identifier", eircode[3:] )
        print("Destination:", key[1])    
        
  else:
    print("Eircode:" , eircode)
    print("Invalid Eircode")

```

```
!wget http://www.gutenberg.org/files/100/100-0.txt

filePath = "/content/100-0.txt"

ShakespeareFile = open(filePath, 'r')
ShakespeareContent = ShakespeareFile.read()
ShakespeareContent = ShakespeareContent.splitlines()
print(ShakespeareContent[0:50])
```

Part 3 Tokenizer 

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer

# setting up tokenizer 
tokenizer = Tokenizer()
# fittign the tokenizer to the Shakespeare text/content
tokenizer.fit_on_texts(ShakespeareContent)

# tokens of shakespeare text/content, tensorflow lemmazises already 
word_index = tokenizer.word_index
print(word_index)
#printed this out later in the file, just so output looks better 
#print("Number of types of tokens:" ,len(word_index))


# to get total number of tokens, we must look at the word counts for each lemma/word 
total_tokens = 0
counts = tokenizer.word_counts

print(counts)

# looping through the dictionary & calculating the values of each key
for x in counts:
  total_tokens += counts[x]

# add a space for readibility
print("")

print("Number of types of tokens:" ,len(word_index))

print("Total number of tokens:" ,total_tokens)

# getting the max key & val 
max_key = None
max_val = None

for key, val in counts.items():
    if max_val is None or val > max_val:
        max_val = val
        max_key = key

print("Most frequent word:", max_key)
print("Number of times used:", max_val)
```

Part 4: NLTK stem 

```
import nltk 
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
#nltk.download('punkt')

# initalize stemmer 
ps = PorterStemmer()
stem_token = []
token = []

# go through each sentence of Shakespeare text/content
for sentence in ShakespeareContent:
   # break into word tokens 
   words = (word_tokenize(sentence))
   for w in words:
     # just store each token, dont stem it
     token.append(w)

# go through each sentence of Shakespeare text/content
for sentence in ShakespeareContent:
   # break into word tokens 
   words = (word_tokenize(sentence))
   for w in words:
     # for each word token, we will stem it 
     stem_token.append(ps.stem(w))

# using set returns the length of unique values of a list
# we can see there is 23031 types after stemming
# roughly 40% decrease
print("No Stemming, number of types of tokens:" , len(set(token)))
print("After Stemming, number of types of tokens:" , len(set(stem_token)))
```
